```python
import os
import pandas as pd
from google_play_scraper import app

print("Libraries imported successfully")
```

```python
# App ID (Swiggy)
APP_ID = "in.swiggy.android"

# Data folder (one level up from notebooks)
DATA_DIR = "../data"
os.makedirs(DATA_DIR, exist_ok=True)

print("Data folder:", os.path.abspath(DATA_DIR))
```

```python
reviews_raw = app(
    APP_ID,
    lang="en",
    country="in"
)

print("Total reviews collected:", len(reviews_raw))
```

```python
from google_play_scraper import reviews, Sort

reviews_raw, _ = reviews(
    APP_ID,
    lang="en",
    country="in",
    sort=Sort.NEWEST,
    count=200
)

print("Total reviews collected:", len(reviews_raw))
```

```python
df = pd.DataFrame(reviews_raw)

print("DataFrame shape:", df.shape)
df.head()
```

```python
df["date"] = pd.to_datetime(df["at"]).dt.date
df[["content", "date"]].head()
```

```python
csv_path = os.path.join(DATA_DIR, "all_reviews.csv")
df.to_csv(csv_path, index=False)

print("✅ CSV saved at:")
print(os.path.abspath(csv_path))
```
