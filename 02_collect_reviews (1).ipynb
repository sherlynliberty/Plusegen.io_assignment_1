```python
import pandas as pd
from google_play_scraper import reviews, Sort

print("Libraries imported successfully")

```

```python
# -----------------------------
# App Configuration
# -----------------------------

APP_NAME = "Swiggy"
APP_ID = "in.swiggy.android"

print("App selected:", APP_NAME)
print("App ID:", APP_ID)
```

```python
# -----------------------------
# Collect reviews from Google Play
# -----------------------------

reviews_raw, continuation_token = reviews(
    APP_ID,
    lang='en',
    country='in',
    sort=Sort.NEWEST,
    count=200
)

print("Total reviews collected:", len(reviews_raw))

```

```python
# -----------------------------
# Convert reviews to DataFrame
# -----------------------------

df = pd.DataFrame(reviews_raw)

print("DataFrame shape:", df.shape)
df.head()
```

```python
# -----------------------------
# Create date column from timestamp
# -----------------------------

df['date'] = pd.to_datetime(df['at']).dt.date

print("Date column created")
df[['content', 'score', 'date']].head()
```

```python
# -----------------------------
# Group reviews by date
# -----------------------------

daily_groups = df.groupby('date')

print("Number of daily batches:", len(daily_groups))
list(daily_groups.groups.keys())[:5]
```

```python
# -----------------------------
# Define data directory
# -----------------------------

DATA_DIR = "data"
```

```python
# -----------------------------
# Save daily batches to /data
# -----------------------------

import os

os.makedirs(DATA_DIR, exist_ok=True)

for date, group in daily_groups:
    file_path = os.path.join(DATA_DIR, f"reviews_{date}.csv")
    group.to_csv(file_path, index=False)

print("Daily batch files saved to:", DATA_DIR)
```
