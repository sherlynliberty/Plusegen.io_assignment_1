```python
import pandas as pd
import os
import re

DATA_DIR = "../data"
csv_path = os.path.join(DATA_DIR, "all_reviews.csv")

df = pd.read_csv(csv_path)
print("CSV loaded:", df.shape)
df.head()

```

```python
def clean_text(text):
    if pd.isna(text):
        return ""
    text = text.lower()
    text = re.sub(r"[^a-z\s]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text
```

```python
df["clean_content"] = df["content"].apply(clean_text)
df["date"] = pd.to_datetime(df["date"])

df[["content", "clean_content", "date"]].head()
```

```python
daily_groups = (
    df.groupby(df["date"].dt.date)
      .agg({
          "clean_content": list,
          "score": "mean",
          "content": "count"
      })
      .reset_index()
      .rename(columns={
          "date": "day",
          "content": "review_count",
          "score": "avg_rating"
      })
)

daily_groups
```

```python
daily_groups["combined_text"] = daily_groups["clean_content"].apply(
    lambda texts: " ".join(texts)
)
```

```python
issue_keywords = {
    "delivery_delay": ["late", "delay", "delayed", "slow"],
    "missing_items": ["missing", "not received", "didnt get"],
    "refund_issues": ["refund", "money back"],
    "app_issues": ["crash", "bug", "error", "app not working"],
    "customer_support": ["support", "help", "customer care"]
}
```

```python
def extract_topics(text):
    found = []
    text = text.lower()
    for topic, keywords in issue_keywords.items():
        if any(k in text for k in keywords):
            found.append(topic)
    return found
```

```python
daily_groups["issues_detected"] = daily_groups["combined_text"].apply(extract_topics)

daily_groups[["day", "issues_detected"]]
```

```python
def sentiment_label(rating):
    if rating >= 4:
        return "positive"
    elif rating == 3:
        return "neutral"
    else:
        return "negative"
```

```python
df["sentiment"] = df["score"].apply(sentiment_label)

df[["score", "sentiment"]].head()
```

```python
sentiment_daily = (
    df.groupby(df["date"].dt.date)
      .agg(
          avg_rating=("score", "mean"),
          review_count=("content", "count"),
          negative_pct=("sentiment", lambda x: (x == "negative").mean() * 100)
      )
      .reset_index()
      .rename(columns={"date": "day"})
)

sentiment_daily
```

```python
def trend_label(row):
    if row["avg_rating"] >= 4 and row["negative_pct"] < 15:
        return "Healthy"
    elif row["avg_rating"] >= 3:
        return "Watch"
    else:
        return "Critical"

sentiment_daily["sentiment_status"] = sentiment_daily.apply(trend_label, axis=1)

sentiment_daily
```

```python
negative_df = df[df["sentiment"] == "negative"].copy()

print("Negative reviews:", negative_df.shape[0])
negative_df[["content", "score"]].head()
```

```python
ISSUE_KEYWORDS = {
    "delivery": ["late", "delay", "delayed", "slow", "waiting"],
    "refund": ["refund", "money", "returned", "return"],
    "support": ["support", "customer care", "executive", "help"],
    "order": ["cancel", "cancelled", "missing", "wrong order"],
    "payment": ["payment", "charged", "upi", "card"]
}
```

```python
def extract_issues(text):
    text = text.lower()
    found = set()

    for issue, keywords in ISSUE_KEYWORDS.items():
        for word in keywords:
            if word in text:
                found.add(issue)

    return list(found)
```

```python
negative_df["issues"] = negative_df["clean_content"].apply(extract_issues)

negative_df[["clean_content", "issues"]].head()
```

```python
issues_exploded = negative_df.explode("issues")

issues_exploded.head()
```

```python
root_causes = (
    issues_exploded["issues"]
    .value_counts()
    .reset_index()
    .rename(columns={"index": "issue", "issues": "count"})
)

root_causes
```

```python
import pandas as pd

df = pd.read_csv("../data/all_reviews.csv")
df["date"] = pd.to_datetime(df["date"])

df = df[["content", "date"]].dropna()
df.head()
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")

embeddings = model.encode(
    df["content"].tolist(),
    show_progress_bar=True
)
```

```python
from sklearn.cluster import AgglomerativeClustering

clusterer = AgglomerativeClustering(
    n_clusters=None,
    distance_threshold=1.2
)

df["topic_cluster"] = clusterer.fit_predict(embeddings)
```

```python
def name_topic(texts):
    return texts[0][:60] + "..."

topic_map = (
    df.groupby("topic_cluster")["content"]
      .apply(list)
      .apply(name_topic)
      .to_dict()
)

df["topic"] = df["topic_cluster"].map(topic_map)

df[["content", "topic"]].head()
```

```python
trend_table = (
    df.groupby(["topic", df["date"].dt.date])
      .size()
      .unstack(fill_value=0)
      .sort_index(axis=1)
)

trend_table.head()
```

```python
trend_table.to_csv("../output/topic_trend_report.csv")
print("‚úÖ Trend report saved to /output/")
```
